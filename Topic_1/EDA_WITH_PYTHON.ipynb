From the output we can see the datatypes stored in the dataframe . Interestingly we can see that there is only one column with a boolean datatype (True or False). We can change the datatype of the Columns by using the broadcasting features in Pandas and the astype attribute

```python
df["Churn"]=df["Churn"].astype("int64") 
df.head()
```

We can see that the Datatype has been changed from boolean to int64 . This is equivalent to the mapping False to 0 and True to 1.

```python
df.describe(include = ["bool","object"])
```

#Indexing
We can index a pandas dataframe in various methods. We can access the series directly by indexing the column or we can use Boolean Indexing.The syntax is df[P(df['Name'])], where P is some logical condition that is checked for each element of the Name column. The result of such indexing is the DataFrame consisting only of the rows that satisfy the P condition on the Name column.

```python
df["Churn"].mean()
```

```python
df.select_dtypes(include=["number"])[df["Churn"] == 1].mean()
```

How much time (on average) do churned users spend on the phone during daytime?

```python
df["Total day minutes"][df["Churn"]==1].mean()
```

What is the maximum length of international calls among loyal users (Churn == 0) who do not have an international plan?

```python
df["Total intl minutes"][(df["Churn"]==0) & (df["International plan"]=="No")].max()
```

DataFrames can be indexed by column name (label) or row name (index) or by the serial number of a row. The loc method is used for indexing by name, while iloc() is used for indexing by number.

```python
df.loc[0:5, ["State", "Churn", "Total day charge"]]
```

```python
df.iloc[0:5, [0,1,2]]
```
